```python
import cv2
import easyocr
```

```python
reader = easyocr.Reader(['en'])
```

```python
def extract_text_from_bbox(image):
    x, y, w, h = 380, 468, 149, 47
    cropped_image = image[y:y+h, x:x+w]
    
    cropped_image_path = 'cropped_1.jpg'
    cv2.imwrite(cropped_image_path, cropped_image)
    # Convert cropped image to grayscale
    gray_image = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)
    
    # Extract text using EasyOCR
    result = reader.readtext(gray_image)
    return result
```

```python
image = cv2.imread('page_1.png')
# Extract text from bounding box region
extracted_text = extract_text_from_bbox(image)

print("Extracted Text:", extracted_text[0][1])
```
